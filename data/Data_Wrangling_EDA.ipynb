```python
import pandas as pd
```

```python
## Loading the datasets

climate_df = pd.read_csv('Climate1985_2020/3WeeksRunningAvg/Canola_TWN_PRA_1985_2020_Climate_MovingAvg_week40.txt',sep='\t')
nvdi_df = pd.read_csv('NDVI_MODIS/3weekMovingAvg/Canola_TWP_MovingAvg.csv')
ecv_smos_df = pd.read_csv('ECV_SMOS2010-2020/TWN_Canola_ECV_SMOS_Weekly_20100101.20201231.csv')
crop_yield_canola_df = pd.read_excel('Crop_Yield/Yield2010_2021_SKMB_Canola_Township_ED.xlsx')
```

```python
## Filtering for years between 2010-2020

climate_df_filtered = climate_df[(climate_df['Year']>=2010) & (climate_df['Year']<=2020)]
nvdi_df_filtered = nvdi_df[(nvdi_df['YEAR']>=2010) & (nvdi_df['YEAR']<=2020)]
ecv_smos_df_filtered = ecv_smos_df[(ecv_smos_df['YEAR']>=2010) & (ecv_smos_df['YEAR']<=2020)]
```

```python
## Renaming the required columns

climate_df_filtered. rename(columns = {'TWN_ID':'TWP_ID', 'Year':'YEAR'}, inplace = True)
```

```python
climate_df_filtered.head(5)
```

```python
nvdi_df_filtered.head(5)
```

```python
climate_nvdi_merged = pd.merge(climate_df_filtered, nvdi_df_filtered, on=['TWP_ID', 'YEAR'], how='inner')
```

```python
climate_nvdi_merged.head(5)
```

```python
crop_yield_canola_col=["TWP_ID","ECODISTRICT_ID","ECOREGION_ID","ECOZONE_ID","ECOPROVINCE_ID"]+[col for col in crop_yield_canola_df if col.startswith("YieldKgAcre")]
crop_yield_canola=crop_yield_canola_df[crop_yield_canola_col]
crop_yield_canola_melt=pd.melt(crop_yield_canola,id_vars=['TWP_ID','ECODISTRICT_ID','ECOREGION_ID','ECOZONE_ID','ECOPROVINCE_ID'],var_name='YEAR',value_name='YieldKgAcre')
crop_yield_canola_melt["YEAR"]=crop_yield_canola_melt["YEAR"].str.slice(start=11)
```

```python
crop_yield_canola_melt.head(10)
```

```python
crop_yield_canola_melt['YEAR'] = crop_yield_canola_melt['YEAR'].astype(int)
```

```python
pd.merge(climate_nvdi_merged, crop_yield_canola_melt, on=['TWP_ID', 'YEAR'], how='inner')
```
